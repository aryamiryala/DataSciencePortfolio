<h3> Step 2: Data Preperation </h3>

- drop any irrelevant columns and rows 
- rename columns 
- feature creation 

```python
#delete any columns I may not need, only columns I don't want is what teams the players played for 
df = df[['Rank', 
    'Player', 
    'Pos', 
    #'Team(s) played for (years)', 
    'Total points',
       'Games  played', 
    'Points  per game[c]', 
    'Field goals Made',
       'Three-pointers Made', 
    'Free  throws made']]
```

```python
#confirm column got dropped
df.shape
```

```python
df.dtypes
```

```python
df.columns
```

```python
#rename columns 
df = df.rename(columns = {'Points  per game[c]': 'Points per Game', 
                    'Total points': 'Total Points', 
                    'Games  played': 'Games Played', 
                    'Free  throws made': 'Free Throws Made'})
```

```python
#confirm everything got renamed correctly 
df.head()
```

```python
#find any missing values in data 
#we can see there are 8 players that have no values for three pointers made 
df.isna().sum()
```

```python
#drop the rows that have no values for three pointers made
df = df.dropna(subset=['Three-pointers Made'])
```

```python
#confirm 8 rows got dropped 
df.shape
```

```python
#confirm no more missing values
df.isna().sum()
```

<h3> Step 3: Feature Understanding </h3> 

- plotting feature distributions 
    - histograms 
    - boxplots
    - KDE

```python
#bar graph showing top positions on all time scoring list
ax = df['Pos'].value_counts().plot(kind='bar', title = "Top Positions on All Time Scoring List")
ax.set_xlabel("Positions")
ax.set_ylabel("Count")
plt.show()
```

```python
#histogram that shows frequency of points per game on all time scoring list 
ax = df['Points per Game'].plot(kind='hist', bins = 20, title = 'Points per Game on All Time Scoring List')
ax.set_xlabel('Points per Game')
plt.show()
```

```python
#same as histogram above expect easier to see with a KDE graph 
ax = df['Points per Game'].plot(kind='kde', title = 'Points per Game on All Time Scoring List')
ax.set_xlabel('Points per Game')
plt.show()
```

<h3> Step 4: Export Cleaned up Data back to CSV</h3>

```python
#convert cleaned up data to new csv file without the row indexes
df.to_csv('TopNBAPlayerPoints_cleaned.csv', index = False)
```
